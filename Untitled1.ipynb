```python
import os
import glob
import re
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import torchvision.transforms as transforms
import torchvision.models as models

# =====================================================
# 1. THE ARCHITECTURE (Transfer Learning via ResNet50)
# =====================================================

class ResNet50Classifier(nn.Module):
    def __init__(self, num_classes=4):
        super().__init__()
        self.resnet = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)

        num_features = self.resnet.fc.in_features

        self.resnet.fc = nn.Sequential(
            nn.Linear(num_features, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(512, 128),
            nn.ReLU(inplace=True),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        return self.resnet(x)


# =================================================
# 2. REAL DATA LOADING FROM DOWNLOADED DIRECTORIES
# =================================================

class BrainTumorDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.class_names = ['glioma', 'meningioma', 'notumor', 'pituitary']
        self.image_paths = []
        self.labels = []

        for label_idx, class_name in enumerate(self.class_names):
            class_folder = os.path.join(root_dir, class_name)
            if os.path.exists(class_folder):
                for img_name in os.listdir(class_folder):
                    self.image_paths.append(os.path.join(class_folder, img_name))
                    self.labels.append(label_idx)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, torch.tensor(label, dtype=torch.long)


# ============
# 3. TRAINING
# ============

if __name__ == "__main__":
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    transform_pipeline = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(15),
        transforms.ColorJitter(brightness=0.1, contrast=0.1),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    base_path = './brain_tumor_data'

    target_path = None
    for root, dirs, files in os.walk(base_path):
        if 'Training' in dirs:
            target_path = os.path.join(root, 'Training')
            break

    if target_path is None:
        raise FileNotFoundError("Could not find 'Training' folder. Please check your unzipped files.")

    print(f"Found real dataset path at: {target_path}")

    train_dataset = BrainTumorDataset(root_dir=target_path, transform=transform_pipeline)

    if len(train_dataset) == 0:
        raise ValueError(f"Dataset is still empty! Checked folder: {target_path}")

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, drop_last=True, num_workers=2, pin_memory=True)
    print(f"Loaded {len(train_dataset)} real brain images for training.")

    NUM_EPOCHS = 30
    model = ResNet50Classifier(num_classes=4).to(device)

    checkpoint_files = glob.glob('resnet_classifier_epoch_*.pth')
    start_epoch = 0
    latest_checkpoint = None

    if checkpoint_files:
        epochs_saved = [int(re.search(r'epoch_(\d+)\.pth', f).group(1)) for f in checkpoint_files if re.search(r'epoch_(\d+)\.pth', f)]
        if epochs_saved:
            start_epoch = max(epochs_saved)
            latest_checkpoint = f'resnet_classifier_epoch_{start_epoch}.pth'

    if latest_checkpoint and os.path.exists(latest_checkpoint):
        model.load_state_dict(torch.load(latest_checkpoint, map_location=device))
        model.to(device)
        print(f"--> [AUTO-RESUME] Loaded weights from ResNet Epoch {start_epoch}.")
    else:
        print("--> Starting fresh from Epoch 1 using ResNet50 Backbone...")

    # موازنةأوزان لتقوية النسب
    class_weights = torch.tensor([2.5, 2.0, 2.0, 1.0]).to(device)
    criterion = nn.CrossEntropyLoss(weight=class_weights)

    optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-2)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=NUM_EPOCHS)

    for _ in range(start_epoch):
        scheduler.step()

    print("Starting Deep Learning Training Loop...")

    for epoch in range(start_epoch, NUM_EPOCHS):
        model.train()
        running_loss = 0.0
        correct_predictions = 0
        total_samples = 0

        for batch_idx, (data, targets) in enumerate(train_loader):
            data, targets = data.to(device), targets.to(device)

            outputs = model(data)
            loss = criterion(outputs, targets)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * data.size(0)
            _, preds = torch.max(outputs, 1)
            correct_predictions += torch.sum(preds == targets.data)
            total_samples += data.size(0)

        epoch_loss = running_loss / total_samples
        epoch_acc = (correct_predictions.double() / total_samples) * 100

        scheduler.step()

        print(f"Epoch {epoch+1}/{NUM_EPOCHS} | Loss: {epoch_loss:.4f} | Accuracy: {epoch_acc:.2f}% | LR: {optimizer.param_groups[0]['lr']:.6f}")
        print("-" * 40)

        if (epoch + 1) % 5 == 0:
            torch.save(model.state_dict(), f'resnet_classifier_epoch_{epoch+1}.pth')
            print(f"--> Checkpoint saved safely: resnet_classifier_epoch_{epoch+1}.pth")

    print("=== Full Training Process Completed Successfully! ===")
```

```python
import os
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import torchvision.transforms as transforms
import torchvision.models as models
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix

class ResNet50Classifier(nn.Module):
    def __init__(self, num_classes=4):
        super().__init__()
        self.resnet = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
        num_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Sequential(
            nn.Linear(num_features, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.3),
            nn.Linear(512, 128),
            nn.ReLU(inplace=True),
            nn.Linear(128, num_classes)
        )
    def forward(self, x): return self.resnet(x)

class BrainTumorDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.class_names = ['glioma', 'meningioma', 'notumor', 'pituitary']
        self.image_paths = []
        self.labels = []
        for label_idx, class_name in enumerate(self.class_names):
            class_folder = os.path.join(root_dir, class_name)
            if os.path.exists(class_folder):
                for img_name in os.listdir(class_folder):
                    self.image_paths.append(os.path.join(class_folder, img_name))
                    self.labels.append(label_idx)
    def __len__(self): return len(self.image_paths)
    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        image = Image.open(img_path).convert('RGB')
        label = self.labels[idx]
        if self.transform: image = self.transform(image)
        return image, torch.tensor(label, dtype=torch.long)

if __name__ == "__main__":
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    transform_pipeline = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    test_dataset = BrainTumorDataset(root_dir='./brain_tumor_data/Testing', transform=transform_pipeline)
    test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

    model = ResNet50Classifier(num_classes=4).to(device)

    # غيري الاعداد حسب اخر دورة!!
    model.load_state_dict(torch.load('resnet_classifier_epoch_15.pth', map_location=device))
    model.eval()

    all_targets, all_predictions = [], []
    print(f"Evaluating model on {len(test_dataset)} NEW images...")

    with torch.no_grad():
        for data, targets in test_loader:
            data = data.to(device)
            outputs = model(data)
            _, predicted = torch.max(outputs, 1)
            all_targets.extend(targets.numpy())
            all_predictions.extend(predicted.cpu().numpy())

    all_targets, all_predictions = np.array(all_targets), np.array(all_predictions)
    class_names = ['Glioma', 'Meningioma', 'No Tumor', 'Pituitary']

    print(f"\n==================================================")
    print(f" MEDICAL EVALUATION REPORT")
    print(f"==================================================")
    print("\n1. Detailed Classification Metrics:")
    print(classification_report(all_targets, all_predictions, target_names=class_names, digits=2))

    print("\n2. Confusion Matrix :")
    cm = confusion_matrix(all_targets, all_predictions)
    print(f"{'True \\ Pred':<12} | {'Glioma':<10} | {'Meningioma':<10} | {'No Tumor':<10} | {'Pituitary':<10}")
    print("-" * 65)
    for i, row in enumerate(cm):
        print(f"{class_names[i]:<12} | {row[0]:<10} | {row[1]:<10} | {row[2]:<10} | {row[3]:<10}")

    print(f"\n==================================================")
```
